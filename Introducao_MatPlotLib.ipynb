```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.arange(6)
print(arr)
print()
plt.plot(arr)
```

```python
arr = np.arange(1, 5) # nesse caso quem varia até 4 é só o y, enquanto o X,quando nao especificado,
#plota do número de y - 1
print(arr)
print()
plt.plot(arr)
```

```python
x = np.arange(0, 10, 0.02)
y = np.sin(x)
plt.plot(x, y)
```

```python
x = np.arange(0, 10, 0.02)
y = np.sin(x)
plt.plot(x, y)
plt.title("Seno") #título
plt.xlabel("X") #nome do eixo x
plt.ylabel("Sen(X)") #nome do eixo y
plt.grid(True) #Cria uma Grade, facilitando a visualização
```
