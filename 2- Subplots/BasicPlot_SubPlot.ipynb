```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
arr_x = np.arange(0, 10, 0.01)
arr_y = np.sin(arr_x)
plt.plot(arr_x, arr_y)
plt.grid(True)
```

* Plotando mais de um gráfico na mesma axis:

```python
arr_y2 = np.cos(arr_x)
plt.plot(arr_x, arr_y, arr_x, arr_y2)

```

```python
#Outra forma
plt.plot(arr_x, arr_y)
plt.plot(arr_x, arr_y2)
```

```python
arr_y3 = np.sin(arr_x) + np.cos(4*arr_x)
plt.plot(arr_x, arr_y, label = 'Sen(x)')
plt.plot(arr_x, arr_y2, label = "Cos(x)")
plt.plot(arr_x, arr_y3, label = "Sen(x) + Cos(3x)")
plt.grid(True)
#Criando uma legenda:
plt.legend(loc = 'lower left')
#definindo os valores dos eixos
plt.xticks(range(0, 11))
plt.yticks([-2,0,2])
```

* **SubPlots:**

```python
plt.figure(figsize = (12,4))  #crio uma figura e passo o tamanho dela com o figsize = (x,y)/ largura, altura
plt.subplot(1,3,1)    #quantidade de linhas, quantidade de colunas, a posição do gráfico
plt.plot(arr_x, arr_y)
plt.subplot(1,3,2)
plt.plot(arr_x, arr_y2, color = 'green')
plt.subplot(1,3,3)
plt.plot(arr_x, arr_y3, color = 'red')
plt.tight_layout() #otimiza os espaçamentos para ter uma melhor visualização
```

```python
arr_y4 = np.cos(arr_x) + np.sin(3 * arr_x)
```

```python
plt.figure(figsize = (8,4))
plt.subplot(2,2,1)
plt.plot(arr_x, arr_y, color = 'blue')
plt.subplot(2,2,2)
plt.plot(arr_x, arr_y2, color = 'green')
plt.subplot(2,2,3)
plt.plot(arr_x, arr_y3, color = 'red')
plt.subplot(2,2,4)
plt.plot(arr_x, arr_y4, color = 'orange')
plt.tight_layout()
```
